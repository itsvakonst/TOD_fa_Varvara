```python
arr = np.random.randint(0, 10, size = (8, 10))
print(arr)
sum_lst = list(np.sum(arr, axis=1))
print(f"–ò–Ω–¥–µ–∫—Å —Å—Ç—Ä–æ–∫–∏ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π: {sum_lst.index(min(sum_lst))}")
print(f"–°—Ç—Ä–æ–∫–∞: {arr[sum_lst.index(min(sum_lst))]}")
```

4. –†–µ—à–∏—Ç—å –º–∞—Ç—Ä–∏—á–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ `A*X*B=-C` - –Ω–∞–π—Ç–∏ –º–∞—Ç—Ä–∏—Ü—É `X`. –ì–¥–µ `A = [[-1, 2, 4], [-3, 1, 2], [-3, 0, 1]]`, `B=[[3, -1], [2, 1]]`, `C=[[7, 21], [11, 8], [8, 4]]`.

```python
A = np.array([[-1, 2, 4], [-3, 1, 2], [-3, 0, 1]])
B = np.array([[3, -1], [2, 1]])
C = np.array([[7, 21], [11, 8], [8, 4]])

B1, C1 = np.linalg.inv(B), -1*C
x = np.linalg.solve(A, C1.dot(B1))
print(f'X = \n{x}')

print(f'\n–ü—Ä–æ–≤–µ—Ä–∫–∞: A * X * B = \n {A.dot(x).dot(B)}')
print(f'\n-C = \n{C1}')
print('–í—Å—ë –≤–µ—Ä–Ω–æ.')
```
