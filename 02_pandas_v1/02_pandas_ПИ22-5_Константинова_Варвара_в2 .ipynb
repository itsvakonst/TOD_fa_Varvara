```python
recipes['name_word_count'] = recipes['name'].str.findall(r'(\w+)').str.len()
recipes
```

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`.
